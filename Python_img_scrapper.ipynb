```python
#library
import os
import time
import requests
from selenium import webdriver

driver_path = "./chromedriver"

def search_and_get_image_url(Search,Drive_Path,No_of_images):
            
    driver = webdriver.Chrome(executable_path = Drive_Path)
    url = "https://www.google.com/search?q={}".format(Search)
    try:
        driver.get(url)
    except:
        print("Url not found")
    time.sleep(1)
    driver.find_element_by_link_text("Images").click()
    
    start = 0
    images = driver.find_elements_by_css_selector('img.Q4LuWd')
    
    while len(images)<=No_of_images:
        driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
        images = driver.find_elements_by_css_selector('img.Q4LuWd')
        try:
            driver.find_element_by_css_selector('input.mye4qd').click()
            print('Found {} Urls looking for more'.format(len(Images)))
        except:
            continue
        
    images_url = set()
    end = len(images)
    for img in images[start:end]:
        try:
            img.click()
        except:
            continue
        time.sleep(2)
        urls=driver.find_elements_by_css_selector('img.n3VNCb')
        for url in urls:
            try:
                if url.get_attribute('src') and 'http' in url.get_attribute('src'):
                    act_url=url.get_attribute('src')
                    images_url.add(act_url)        
                    if len(images_url)>=No_of_images:
                        driver.close()
                        return images_url
            except Exception:
                print("Link Not Found")
        
        
def save_url(Search_term,DrivePath,NoOfImages,target_folder = './images'):
    searched_urls = search_and_get_image_url(Search = Search_term, Drive_Path = DrivePath,No_of_images = NoOfImages)
    
    target_path = os.path.join(target_folder,Search_term)
    print(target_path)
    
    if not os.path.exists(target_path):
        os.makedirs(target_path)

    counter=1
    for img_url in searched_urls:
        url_saving_folder(target_path,img_url,counter)
        counter+=1
        
        
        
def url_saving_folder(folder_path,url,counter):
    try:
        img_cont = requests.get(url).content
    except Exception:
        print("URL not Loading")
    try:
        f = open(os.path.join(folder_path,'jpg'+"_"+str(counter)+'.jpeg'),'wb')
        f.write(img_cont)
        f.close()
        print("Saved Image with url - {}".format(url))
    except Exception:
        print("path not found")
```

```python
search = str(input("Enter the searched term whose images needed to be downloaded - "))
no_of_images = int(input("Enter no of images to be downloaded - "))

save_url(Search_term = search, DrivePath = driver_path,NoOfImages = no_of_images)
```
