```
#2.1.2
def filter_long_words(list_arg,n):
    if type(list_arg)==list:
        return [ele for ele in list_arg if len(ele)>n]
    else:
        return "Function only accepts list of words as an input"
#Example:-
filter_long_words(['John','Harry','Alison','Christopher'],4)
```
